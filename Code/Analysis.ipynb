```python
from google_play_scraper import reviews, Sort
import pandas as pd

results, _ = reviews(
    'com.supercell.clashroyale',   # Clash Royale package name
    lang='en',                     # Language
    country='us',                  # Country
    sort=Sort.NEWEST,              # Sort by newest reviews
    count=6000                     # Number of reviews to scrape
)

play_df = pd.DataFrame(results)

# Keep only useful columns
play_df = play_df[['content', 'score', 'at']]
play_df.columns = ['text', 'rating', 'date']

play_df.head(), play_df.shape
```

```python
import re

def clean_text(text):
    text=str(text)
    text=re.sub(r"http\S+|www\S+", "", text)
    text=re.sub(r"&lt;|&gt;|&amp;", " ", text)
    text=re.sub(r"\s+", " ", text)
    text=text.strip()
    return text

Clash_Royal_Reviews['cleaned_text']=Clash_Royal_Reviews['text'].apply(clean_text)
Clash_Royal_Reviews["word_count"] = Clash_Royal_Reviews["cleaned_text"].str.split().str.len()
Clash_Royal_Reviews["text_length"] = Clash_Royal_Reviews["cleaned_text"].str.len()
Clash_Royal_Reviews = Clash_Royal_Reviews[Clash_Royal_Reviews["word_count"] > 2]
```
